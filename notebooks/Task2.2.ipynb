#### Convert Year to Datetime Format in GDP Data

```python
import pandas as pd

# Load economic data with 'Year' instead of full date
economic_data = pd.read_csv('../Data/All_economic_data.csv')

# Convert 'Year' to a datetime format with January 1 as the date
economic_data['Date'] = pd.to_datetime(economic_data['date'], format='%Y')

# Sort by country and date, then calculate the GDP growth rate
economic_data.sort_values(['country', 'date'], inplace=True)
economic_data['GDP_Growth_Rate'] = economic_data.groupby('country')['GDP'].pct_change() * 100

# Drop the original 'Year' column as it's now redundant
economic_data.drop(columns=['date'], inplace=True)

display(economic_data.head(100))
```

#### Prepare Oil Prices Data

```python
# Load oil prices data, ensuring it has a 'Year' or 'Date' column
import pandas as pd

# Load oil prices data
oil_data = pd.read_csv('../Data/BrentOilPrices.csv')

oil_data['Date'] = pd.to_datetime(oil_data['Date'])

# Extract the year from the oil price 'Date'
oil_data['Date'] = oil_data['Date']

# Calculate the annual average oil price
annual_oil_data = oil_data.groupby('Date')['Price'].mean().reset_index()
annual_oil_data.rename(columns={'Price': 'Average_Oil_Price'}, inplace=True)
display(annual_oil_data.head())
```

```python
# Merge the GDP data with oil prices data on 'Date'
combined_data = pd.merge(economic_data, oil_data, on='Date')
combined_data.to_csv('../Data/combined_gdp_price.csv')

# Check combined data
display(combined_data.head(10))
```

```python
# Calculate correlation between GDP Growth Rate and Average Oil Price for each country
correlations = combined_data.groupby('country').apply(
    lambda df: df['GDP'].corr(df['Price'])
)

# Display the correlation results
print("Correlation between GDP growth rate and average oil price:")
display(correlations)
```
